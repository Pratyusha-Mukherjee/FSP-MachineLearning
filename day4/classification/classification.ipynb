### Logistic function in linear regression is  a type of sigmoid, a class of functions with the same specific properties. Sigmoid is a mathematical function that takes any real number and maps the probability between 1 and 0.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def derivative(x, step):
    return (sigmoid(x+step) - sigmoid(x))/step
```

```python
x=np.linspace(-5,5,1000)
y1=sigmoid(x)
y2=derivative(x, 0.0000000000001)
plt.plot(x,y2, label='Derivative')
plt.plot(x,y1, label='Sigmoid')
plt.legend()
plt.grid()
plt.show()
```
