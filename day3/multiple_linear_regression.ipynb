```python
fig,ax=plt.subplots( nrows=1,ncols=2,figsize=(10,8))
ax[0].set_xlabel('crim')
ax[0].set_ylabel('medv')
ax[0].set_title("Scatter plot crim vs medv")
ax[0].scatter(df['crim'],df['medv'],marker='o',color='green')



ax[1].set_title("Histogram crim vs medv")
ax[1].set_xlabel('crim')
ax[1].set_ylabel('medv')
ax[1].hist(df['crim'],color='blue')


plt.show()
```

```python
#reading thr value of the zn
df['zn'].head(5)
```

```python
values=['zero','non-zero']
frequencies = [df[df['zn']==0]['medv'].mean(),
              df[df['zn']!=0]['medv'].mean() ]
sns.barplot(x=values,y=frequencies)
plt.title("Bargraph")
plt.xlabel('values')
plt.ylabel('frequencies')
```

```python
plt.scatter(df['indus'],df['medv'],color='red')
plt.xlabel("Industry")
plt.ylabel("Price")
plt.title("Scatter plot upon indus")
plt.show()
```

```python
df['chas'].value_counts()
```

```python
value=['Zero', 'One']
frequencies=[df[df['chas']==0].shape[0], df[df['chas']==1].shape[0]]
fig=plt.figure(figsize=(10,5))
ax=fig.add_subplot(1,2,1)
plt.title('Pie Chart')
plt.pie(frequencies, labels=value, autopct='%1.1f%%', shadow=False,explode=(0.1,0))
fig.add_subplot(1,2,2)
plt.title('Bar Chart')
plt.bar(value, frequencies, color='blue')
plt.show()

```

```python
plt.scatter(df['nox'],df['medv'],color='grey')
plt.xlabel("Nitrogen oxide")
plt.ylabel("Median house value")
plt.grid()
plt.show()
```

```python
#plotting of RH
fig=plt.figure(figsize=(10,5))
fig.add_subplot(1,2,1)
plt.scatter(df['rm'],df['medv'],color='green')
plt.title('scatter plotting of room with medv')
plt.xlabel('Room')
plt.ylabel('Medv')


fig.add_subplot(1,2,2)
sns.boxplot(y=df['rm'],color='yellow')
plt.show()
```
