```python
def get_corelated_data(corrdata, threshold):
    feature = []
    value = []
    for index in corrdata.index:
        if abs(corrdata.loc[index]) > threshold:
            feature.append(index)
            value.append(corrdata.loc[index])
    
    df = pd.DataFrame(data=value, index=feature, columns=['corr_value'])
    return df
```

```python
threshold = 0.50
corr_value = get_corelated_data(cormat['medv'], threshold)
corr_value
```

```python
corelated_data=df[corr_value.index]
corelated_data
```

```python
sns.pairplot(corelated_data)
```

```python
x = corelated_data.drop(labels = ['medv'], axis = 1)
y = corelated_data['medv']
```
