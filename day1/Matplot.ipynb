```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
plt.grid()
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
temp = [32,36,38,34,35,40,42]
plt.plot(x,temp)
# plt.show()
plt.grid()
```

```python
x = [1,2,3,4,5,6,7]
temp = [32,36,38,34,35,40,42]
plt.title("Climate")
plt.plot(x,temp)
plt.xlabel("Day")
plt.grid()
plt.ylabel("Temp")
```

```python
x = [1,2,3,4,5,6,7]
temp = [32,36,38,34,35,40,42]
plt.title("Climate")
plt.plot(x,temp,marker="o",color="red",linestyle='dashed')
plt.xlabel("Day")
plt.grid()
plt.ylabel("Temp")
```

```python
company = ['Relince','Adani','Tata']
revenue=[110,120,90]

plt.title('company revenue')

xpos = np.arange(len(company))
plt.xticks(xpos,company)

plt.bar(xpos,revenue,width=0.2,color="red",label="revenue")
plt.legend()
plt.show()
```

```python
x=["USA", "China", "Japan", "Australia"]
G=[40,40,20,18]
S=[44,27,12,19]
B=[42,24,13,16]
b_bronze=list(np.add(G,S))
plt.bar(x,G,width=0.4,label="Gold",color="yellow")
plt.bar(x,S,0.4,bottom=G,label="Silver",color="Gray")
plt.bar(x,S,0.4,bottom=b_bronze,label="Bronze",color="Red")
plt.show()
```

```python
m = ['Apple','Realme','Oppo','Vivo','Samsung']
share = [500,400,300,200,100]
plt.pie(share,labels=m,radius=0.75,shadow=True)
plt.show()
```

```python
m = ['Apple','Realme','Oppo','Vivo','Samsung']
share = [500,400,300,200,100]
plt.pie(share,labels=m,radius=0.75,shadow=True,autopct='%1.1f%%')
plt.show()
```

```python
m = ['Apple','Realme','Oppo','Vivo','Samsung']
share = [500,400,300,200,100]
plt.pie(share,labels=m,radius=0.75,shadow=True,counterclock=False,autopct='%1.1f%%',
           explode=[0,0.1,0.1,0.1,0.1])
plt.show()
```
