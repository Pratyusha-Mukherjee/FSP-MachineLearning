Write a code to find the pokemon that is clearly outliers in terms of attack/defense

```python

poke['Attack'] = poke['Attack'].astype(float)
poke['Defense'] = poke['Defense'].astype(float)

poke['attack_defense'] = poke['Attack'] + poke['Defense']

outliars = poke[poke['attack_defense'] > 250]

plt.figure(figsize=(10,10))
plt.scatter(poke['Attack'], poke['Defense'], c='r')
plt.scatter(outliars['Attack'], outliars['Defense'], c='b')
plt.show()
```
