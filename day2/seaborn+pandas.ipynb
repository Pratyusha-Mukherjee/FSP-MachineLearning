write a code to plot relationship between total bill and tip

```python
sns.relplot(x="total_bill", y="tip", data=tp)
```

write a code to check the highest tip given by male or female

```python
sns.relplot(x="total_bill", y="tip", hue="sex",data=tp)
```

Write a code to count the total no of male and female in tips dataset and visualize also

```python
tp['sex'].value_counts()
```

```python
sns.countplot(x='sex',data=tp)
```

write a code to count the day in the tips dataset and also visualize the day of the tip date

```python
tp['day'].value_counts()
```

```python
color = ['Pink','Orange','Yellow','Purple']
sns.countplot(x='day',data=tp,palette=color)
```
